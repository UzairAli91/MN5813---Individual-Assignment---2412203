Auction dynamics often exhibit unpredictable patterns and trends. Some players spark a competitive bidding war amongst franchises, driving their winning bids significantly higher. To better understand these patterns and trends, players must be categorised into specific labels that reflect their market value. For instance, a player sold at an exceptionally high price must not be grouped with the one which is sold closer to their base price. Therefore, players need to be assigned to a certain auction category to illustrate their relative worth and value <br>

This study classifies players into three auction categories: **"Platinum", "Gold" and "Silver"**. <br> 

1. **Platinum:**  represents the most premium players attracting the highest bids. <br>

2. **Gold:** reflects expensive but do not reach the premium level of Platinum players. <br>

3. **Silver:** consists of players sold at prices closer to their base price. <br>

The classification criteria are as follows: <br>

1. Players with winning bids between INR 0.0 - 4.9 Cr. are categorised as **"Silver"**. <br>

2. Players with winning bids between INR 5.0 - 14.9 Cr. are categorised as **"Gold"**. <br>

3. Players with winning bids of INR 15.0 Cr. or more are categorised as **"Platinum"**. <br>

It is important to note that **"Retained"** and **"Unsold"** players are not associated with a winning bid. For these players the new column will display **"None"** as their auction category.

This objective deep dives into the spending patterns and behaviours of IPL franchises across the 2024 and 2025 auctions. By visualising team’s expenditures, the study seeks to shed light on which teams made the topmost financial investments and how spendings differed across both the auction categories and years. ### Visualisation 1: Team Spendings Across Auctions (2024 vs 2025)

```python
#To filter out data for sold players, following code will be used
sold_players = merged_df[merged_df['Player Status'] == 'Sold']

#To define spending by each team separately for 2024 and 2025, following codes will be used
def total_spending_by_team(year):
    df_year = sold_players[sold_players['Year'] == year]
    spending = df_year.groupby('Team')['Winning Bid (₹ Cr.)'].sum().reset_index()
    spending = spending.sort_values(by = 'Winning Bid (₹ Cr.)', ascending = False) 

#To plot the visulisation and set labels, following codes will be used
    plt.figure(figsize=(10, 6))
    sns.barplot(
        data = spending,
        x = 'Winning Bid (₹ Cr.)',
        y = 'Team',
        palette = 'viridis'
    )
    plt.title(f'Team Spendings ({year})', fontsize = 16)
    plt.xlabel('Total Spending (₹ Cr.)')
    plt.ylabel('Team')
    filename = (f'Team_Spendings_{year}.png')
    plt.savefig(filename, dpi=300, bbox_inches='tight') 
    plt.show()

#To plot visualisation for 2024
total_spending_by_team(2024)

#To plot visualisation for 2025
total_spending_by_team(2025) 
```

1. McKinney (2012) explains that **"figsize"** sets the figure a certain size and aspect ratio. It is defined as a tuple indicating the desired dimensions. <br>
2. According to McKinney (2012) the **"groupby"** function allows to slice and dice, and summarise data in a natural way. It serves as the foundation for creating Pivot tables in Python using Pandas. <br>
3. McKinney (2012) states that **"xlabel"** is used to label the X axis, **"ylabel"** is used to label the Y axis and **"title"** assigns a title to the subplot.

**Visualisation 1 - Analysis:** <br>
A horizontal bar chart was used to effectively compare individual team spending while providing clear labelling and readability for categorical data. <br>

### Visualisation 2: Most Expensive Player by Team and Player Role (2024 vs 2025)

```python
#To filter out data for most expensive player by team, following code will be used
def most_expensive_player_by_team(year):  
    df_year = merged_df[(merged_df['Year'] == year) & (merged_df['Player Status'] == 'Sold')]  

#To find out the most expenive purchase by each team, following code will be used
    expensive_players = (
        df_year.loc[df_year.groupby('Team')['Winning Bid (₹ Cr.)'].idxmax()].reset_index(drop = True)
    )  

#To combine player names and roles, following code will be used
    expensive_players['Player Info'] = expensive_players['Player'] + ' (' + expensive_players['Player Role'] + ')' 

#To sort by winning bid, following code will be used
    expensive_players = expensive_players.sort_values(by = 'Winning Bid (₹ Cr.)', ascending = False) 

#To plot the visualisation and set labels, following codes will be used
    plt.figure(figsize=(14,10))
    sns.barplot(
        data = expensive_players,
        x = 'Winning Bid (₹ Cr.)',
        y = 'Team',
        hue = 'Player Info',
        dodge = False,
        palette = 'cool'
    )
    plt.title(f'Most Expensive Player by Team and Player Role ({year})', fontsize = 16)
    plt.xlabel('Winning Bid (₹ Cr.)')
    plt.ylabel('Team')
    plt.legend(title = 'Player Name & Role', bbox_to_anchor = (1.05, 1), loc = 'upper left')
    filename = (f'Most_Expensive_Player_{year}.png')
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    plt.show()

#To plot visualisation for 2024
most_expensive_player_by_team(2024) 

#To plot visualisation for 2025
most_expensive_player_by_team(2025) 
```

1. McKinney (2012) describes **"loc"** as a parameter in Matplotlib specifying where to exactly position the plot legend. <br>
2. McKinney (2012) mentions that the data is sorted in ascending order by default, but sorting in descending order can be achieved by using **"ascending = False"**.
